```python
import matplotlib.pyplot as plt
import numpy as np

def deriv(f,ip): #calculates the derivative of function f at the initial point ip
        return (f(ip+10**(-5))-f(ip))/(10**(-5)) #returns the approximation of the derivative

x1=ip-lr*deriv(f,ip) #sets x1 initially to a value slightly smaller than the initial point 
x0=ip #sets x0 equal to the initial point so that it can be changed within the loop

xc=[x1]
yc=[f(x1)] 

def grad(f, lr, ip): #defines the function and calls the inputed function, the learning rate, and initial point
    x1=ip-lr*deriv(f,ip)
    x0=ip
    for i in range(100): 
        xc.append(x1) 
        yc.append(f(x1))
        if deriv(f,x1)<0.001: #if the derivative is adequately close to 0
            break
        elif deriv(f,x1)<deriv(f,x0):
            x0=x1 #shifts x0 smaller to the x1 value
            x1=x1-lr*deriv(f,x1) #redefines x1 to a slighly smaller value
    plot_range=np.linspace(x1-0.5,x1+0.5,100) #defines the x-value range of the graph 
    function_range=[f(i) for i in plot_range] #defines the y-value range of the graph
    plt.plot(plot_range, function_range) #plots the graph
    plt.scatter(x1,f(x1)) #plots the minimum point found using loop
    return round(xc[-1],3), round(yc[-1],3) #returns the final point found in loop, which is closest to minimum
```

```python
def f(x):
    return x**2
ip=5
lr=0.1

print("The minimum of this function is approximately", grad(f,lr,ip))
```
