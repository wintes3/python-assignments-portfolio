```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean=171
variance=7.1
#defines the function for calculating the probability density of a number 
def normal_density(mean,variance,x):
    prob_density=np.exp(-((x-mean)**2)/(2*variance**2))/(np.sqrt(2*np.pi*variance**2))
    return prob_density
#plots the function for a variety of x-values around the mean
x_values=np.linspace(mean-variance,mean+variance) #generates x-values around mean
plt.plot(x_values, normal_density(mean,variance,x_values))
plt.title("Normal Density")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
#defines the function for calculating the probability a value will fall within an interval
def integration(mean,variance,a,b): 
    n=100000 #number of values within the interval (greater number means more accuracy)
    xi=a
    delta_x=(b-a)/n 
    probability=0
    while xi<=b: #repeats until xi equals b
        probability=probability+normal_density(mean,variance,xi)*delta_x #sums all probability densities
        xi=xi+delta_x #moves xi one step
    return probability
print(integration(171,7.1,162,190))
```
